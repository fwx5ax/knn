**Q0.**
1. What is the difference between regression and classification?

The difference between regression and classification is that regression predicts values based on input data and classification predicts categories or labels based on input data. If we want to predict the price of something, we would use regression. If we want to determine if an athlete plays offense or defense based on their data, we would use classification. Regression aims to predict a specific value of a variable based on the data, while classification labels the data using pre-existing classes/labels/groups.

2. What is a confusion table? What does it help us understand about a model's performance?

A confusion table is used after a binary classification model is used to evaluate how well the model is predicting/assigned the labels. The table has 4 sections showing true positives, false positives, true negatives, and false negatives. We can see how many times the model correctly predicted values and how many times it incorrectly predicted values. In some cases a false negative is worse than a false positive (or vice versa) allowing us to see if the model's errors are meaningful.

3. What does the SSE quantify about a particular model?

The SSE (sum of squared errors) of a model gives quantifies the total deviation of a model's predictions from the actual values. It measures the overall error in a regression model by summing the squared differences between the observed values and the predicted values. It can tell us about the model's fit - the lower the SSE, the closer the model's predictions are to the actual values. It also tells us about error distribution because larger errors (or outliers) affect the SSE more. When we are comparing models with similar complexity, you can compare the models

4. What are overfitting and underfitting?
Overfitting happens when a model is too complex and fits the data too closely. It may be accurate for the current dataset but struggles to work with other data for predictions because it is too specific to the original data. If a model has too many parameter or layers, it might be overfitting. Underfitting is the opposite problem where a model is too simple that it does not fully capture the relationships in the data. An underfit model has low accuracy in general.
